```python
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
ds = pd.read_csv('heart.csv')
ds.head()
```

```python
ds2 = ds[['age', 'target']]
sns_plot = sns.catplot(x="age", hue="target", data=ds2, kind="count");   
sns_plot.set_axis_labels('Age', 'Number of people')
sns_plot.fig.set_figwidth(18);
sns_plot.fig.set_figheight(8);
sns_plot.savefig("img/output1.png", dpi = 1000, transparent = True)
```

```python
ds2 = ds[['sex', 'target']]
sns_plot = sns.catplot(x="sex", hue="target", data=ds2, kind="count");   
sns_plot.set_axis_labels('Gender', 'Number of people')
sns_plot.fig.set_figwidth(18);
sns_plot.fig.set_figheight(8);
sns_plot.savefig("img/output2.png", dpi = 1000, transparent = True)
```
